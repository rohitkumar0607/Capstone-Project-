```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
from selenium.common.exceptions import NoSuchElementException,StaleElementReferenceException
```

```python
def create_and_configure_driver():
    # Create a WebDriver instance
    driver = webdriver.Chrome(r'chromedriver.exe')
    

    # Open booking.com
    driver.get("https://www.booking.com/index.en-gb.html?aid=397646&label=yho748jc-1DCAEoggI46AdIM1gDaGyIAQGYAQm4ARfIAQzYAQPoAQGIAgGoAgO4Ap3Y96YGwAIB0gIkY2NlMDIyZmItZDdlZi00MWVkLWE5YzUtNmIzNmE1NWI3MDU32AIE4AIB&sid=0cb28093647b47834b4f3a02e11acd13&keep_landing=1&sb_price_type=total&")
    time.sleep(2)

    try:
        # Clicking on close to close popup
        dialog_box_close = driver.find_element(By.XPATH, '/html/body/div[17]/div/div/div/div[1]/div[1]/div/button/span/span')
        dialog_box_close.click()
        time.sleep(1)

        # Clicking on cities option
        click_on_cities = driver.find_element(By.XPATH, '//li[@class="bui-tab__item"][2]')
        click_on_cities.click()
    except Exception as e:
        print(f"An error occurred: {str(e)}")

    return driver
```

```python
# Call the function to create and configure the driver
driver = create_and_configure_driver()
```

```python
# Define your lists before using the function

roomtype= []
beds = []
prices= []
cities = []
location = []
hotellinks=[]
cityname=[]
```

```python
def all_data(driver, num_clicks):
    for _ in range(num_clicks):
        

        try:
            room=driver.find_elements(By.XPATH,'//span[@class="df597226dd"]')
            for r in room:
                roomtype.append(r.text)
        except (NoSuchElementException,StaleElementReferenceException) as e:
            print('Exception raised:',e)
            roomtype.append('NaN')
            pass

        try:
            bed=driver.find_elements(By.XPATH,'//div[@class="cb5b4b68a4"]')
            for b in bed:
                beds.append(b.text)
        except (NoSuchElementException,StaleElementReferenceException) as e:
            print('Exception raised:',e)
            beds.append('NaN')
            pass

        try:
            price=driver.find_elements(By.XPATH,'//span[@class="f6431b446c fbd1d3018c e729ed5ab6"]')
            for p in price:
                prices.append(p.text)
        except (NoSuchElementException,StaleElementReferenceException) as e:
            print('Exception raised:',e)
            prices.append('NaN')
            pass

        try:
            city=driver.find_elements(By.XPATH,'//a[@class="a83ed08757 f88a5204c2 e0e11a8307 b98133fb50"]/span/span[1]')
            for c in city:
                cities.append(c.text)
        except (NoSuchElementException,StaleElementReferenceException) as e:
            print('Exception raised:',e)
            cities.append('NaN')
            pass

        
        try:
            url_1 = driver.find_elements(By.XPATH,'//h3[@class="a4225678b2"]/a')
            for i in url_1:
                hotellinks.append(i.get_attribute('href'))
        except (NoSuchElementException,StaleElementReferenceException) as e:
            print('Exception raised:',e)
            hotellinks.append('NaN')
            pass

        # Click on the "Next" button to navigate to the next page
        try:
            click_on_nextbutton = driver.find_element(By.XPATH, '//div[@class="b16a89683f cab1524053"]')
            click_on_nextbutton.click()
        except NoSuchElementException:
            # Handle the case when the "Next" button is not found (end of pages)
            break
        time.sleep(1)    
```

```python
# Specify the number of pages to scrape
num_clicks= 6

# Call the function to extract hotel links
all_data(driver, num_clicks)
```

```python
starrating=[]
breakfast=[]
numberofreviews = []
reviewscore = []
Hotel=[]
driver = webdriver.Chrome(r'chromedriver.exe')
for i in hotellinks:
    driver.get(i)
    try:
        hotels=driver.find_element(By.XPATH,'//h2[@class="d2fee87262 pp-header__title"]')
        Hotel.append(hotels.text)
    except (NoSuchElementException,StaleElementReferenceException) as e:
        print('Exception raised:',e)
        Hotel.append('NaN')
    
    try:
        starratings = driver.find_elements(By.XPATH, '//span[@class="a455730030 e23c0b1d74"]/span')
        count = 0  # Initialize a count variable
        for star in starratings:
            count += 1  # Increment the count
        starrating.append(count)  # Append the count to the list
    except (NoSuchElementException, StaleElementReferenceException) as e:
        print('Exception raised:', e)
        starrating = ['NaN']
        pass
    
    try:
        numberofreview=driver.find_element(By.XPATH,'//span[@class="a3b8729ab1 f45d8e4c32 d935416c47"]')
        numberofreviews.append(numberofreview.text)
    except (NoSuchElementException,StaleElementReferenceException) as e:
        print('Exception raised:',e)
        numberofreviews.append('NaN')
        pass

    try:
        reviewscores=driver.find_element(By.XPATH,'//div[@class="aca0ade214 aaf30230d9 e1ffac4e41 e7d9f93f4d d79e71457a d5fd510f01 dc7f26e57f"]/div[1]')
        reviewscore.append(reviewscores.text)
    except (NoSuchElementException,StaleElementReferenceException) as e:
        print('Exception raised:',e)
        reviewscore.append('NaN')
        pass

    try:
        lm=driver.find_element(By.XPATH,'//span[@class="ph-item-copy-breakfast-option"]')
        breakfast.append(lm.text)
         
    except (NoSuchElementException,StaleElementReferenceException) as e:
        print('Exception raised:',e)
        breakfast.append('No Breakfast')
```

```python
# Split each city and get the second element (index 1) for city if it exists, otherwise 'NaN'
second_elements = [city.split(',')[1] if len(city.split(',')) > 1 else 'NaN' for city in cities]
for element in second_elements:
    cityname.append(element)
# Split each city and get the first element (index 0) for location if it exists, otherwise 'NaN'    
for locatin in cities:
    loc = locatin.split(',')[0] if len(locatin.split(',')) > 0 else 'NaN'
    location.append(loc)    
```
