#### 3) Write a query to count the occurrences of each breakfast cuisine type.

```python
sql_queries = [
    "SELECT 'Breakfast_Cuisine1_Continental' AS Cuisine_Type, COUNT(*) FROM DataTable_2 WHERE Breakfast_Cuisine1_Continental = 'yes'",
    "SELECT 'Breakfast_Cuisine2_Vegetarian' AS Cuisine_Type, COUNT(*) FROM DataTable_2 WHERE Breakfast_Cuisine2_Vegetarian = 'yes'",
    "SELECT 'Breakfast_Cuisine3_American' AS Cuisine_Type, COUNT(*) FROM DataTable_2 WHERE Breakfast_Cuisine3_American = 'yes'"
]

# Initialize an empty dictionary to store the results
results = {}

# Execute the SQL queries and store the results in the dictionary
for query in sql_queries:
    cursor.execute(query)
    row = cursor.fetchone()
    cuisine_type = row[0]
    count = row[1]
    results[cuisine_type] = count

# Print the results
for cuisine_type, count in results.items():
    print(f"{cuisine_type}: {count}")
```
